```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np

numeros1=[str(i) for i in range(1,21)]
numeros2=[round(rd.uniform(0,6),2) for i in numeros1]
print(numeros1,"\n",numeros2,"\n",len(numeros2))

media=np.mean(numeros2)
mediana=np.median(numeros2)
desviacion=np.std(numeros2)

print("media: ",media,"\nmediana: ",mediana,"\ndesviacion: ",desviacion)

plt.bar(numeros1,numeros2)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
