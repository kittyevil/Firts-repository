```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
f=6
c=6
cantidadcarros=0
carros=[]
distancia=[]
matriz=np.random.randint(0,2,size=(f,c))
print(matriz)
for i in range(len(matriz)):
    for e in range(len(matriz[i])):
        if matriz[i,e] == 1:
            matriz[i,e]=rd.randint(100,999)
            cantidadcarros+=1
            carros.append(str(matriz[i,e]))
            distancia.append(str(i+e))
print(carros,distancia,"\n \n",matriz)
plt.plot(carros,distancia,linestyle="--")
plt.show()
```
