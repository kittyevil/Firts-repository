```python
import random as rd
import matplotlib.pyplot as plt


normal=[rd.randrange(-100,100) for i in range(100)]
cubo=[e**3 for e in normal]
# grafica normal
prgrafica=plt
prgrafica.subplot(3,1,1)
prgrafica.scatter(normal, cubo,)
plt.ylabel("x**3")
plt.xlabel("x")

# grafica barra
prgrafica=plt
prgrafica.subplot(3,1,2)
prgrafica.bar(normal, cubo,)
# plt.ylim(-100,10)
print(normal[:5],cubo[:5])
plt.ylabel("x**3")
plt.xlabel("x")

#grafica linea
prgrafica=plt
prgrafica.subplot(3,1,3)
prgrafica.plot(normal, cubo, linestyle = "--")
plt.ylabel("x**3")
plt.xlabel("x")
# plt.ylim(-100000,100000)
plt.show()
```
