```python
import random as rd
import matplotlib.pyplot as plt
# a = [uno, dos, tre]
# juan, andres, alan,william
listanombres=input("ingrese lista de nombres")
print(listanombres)
listanombres=listanombres.split(",")
listaedades=[]
print(listanombres,len(listanombres))
for i in range(len(listanombres)):
    #print(i)
    e = rd.randint(0,30)
    listaedades.append(e)
print(listaedades,listanombres)
# grafica normal
prgrafica=plt
prgrafica.subplot(3,1,1)
prgrafica.scatter(listanombres, listaedades, label = "nombres")
plt.ylabel("edades")
plt.xlabel("nombre")
prgrafica.legend(loc = 'upper right')
print(listaedades,listanombres)
# grafica barra
prgrafica=plt
prgrafica.subplot(3,1,2)
prgrafica.bar(listanombres, listaedades, label = "nombres")
plt.ylabel("edades")
plt.xlabel("nombre")
prgrafica.legend(loc = 'upper right')
#grafica linea
prgrafica=plt
prgrafica.subplot(3,1,3)
prgrafica.plot(listanombres, listaedades, linestyle = "--")
plt.ylabel("edades")
plt.xlabel("nombre")
prgrafica.legend(loc = 'upper right')
plt.show()
```
